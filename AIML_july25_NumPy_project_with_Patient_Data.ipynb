###  Loading Dataset

```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/sarwansingh/Python/refs/heads/master/ClassExamples/data/patients.csv'
pat = np.genfromtxt(url, delimiter=',', dtype='object', skip_header=1)
# genfromtxt (fname, dtype=<class 'float'>, comments='#', delimiter=None, skip_header=0,
# skip_footer=0, converters=None, missing_values=None, filling_values=None, usecols=None, names=None,
# excludelist=None, deletechars=" !#$%&'()*+,-./:;<=>...", replace_space='_', autostrip=False, case_sensitive=True,
# defaultfmt='f%i', unpack=None, usemask=False, loose=True, invalid_raise=True, max_rows=None, encoding='bytes')
# names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
```

```python
# see the first 10 rows
pat[:10]
# patientID,  age, bloodpressure ,heartrate, body temp, oxygen level
```

```python
#remove the first column
pat = pat[:,1:]
pat[:10]
```

```python
# convert all object datatype in int/float
pat = pat [:, :].astype(float)
pat [:10]
```

```python
a1 = np.array ( [[ 40,159] , [65,200] , [50,135], [45,150], [60,175], [50,138]])
# np.sort(a1, axis=1)
a1.sort(axis=0)
print(a1)
```

```python
corr, p_value = pearsonr(a1[:, 0], a1[:, 1])
print(f'Pearson correlation coefficient: {corr:.2f}')
print(f'P-value: {p_value:.4f}')
```

```python
a2 = np.array ( [[ 40,159] , [65,280] , [50,135], [45,150], [60,175], [50,138]])
# np.sort(a1, axis=1)
a2.sort(axis=0)
print(a2)
corr, p_value = pearsonr(a2[:, 0], a2[:, 1])
print(f'Pearson correlation coefficient: {corr:.2f}')
print(f'P-value: {p_value:.4f}')
```

```python
import statsmodels.api as sm

X = a1[:, 0]
y = a1[:, 1]
X = sm.add_constant(X)  # Adds intercept term

model = sm.OLS(y, X).fit()
print(model.summary())
```
